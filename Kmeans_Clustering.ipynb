```
#import library for clustering
from sklearn.cluster import KMeans
```

```
#create Kmeans clustering model with 2 cluster
kmeans=KMeans(n_clusters=2)
```

```
#fit the model from all date except variable 'private'
kmeans.fit(df.drop('Private',axis=1))
```

```
#find cluster center vectors
kmeans.cluster_centers_
```

```
#Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.
def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
```

```
df['Cluster'] = df['Private'].apply(converter)
```

```
df.head()
```

```
#create the confusion matrix
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```

```
# add prediction label to df
df['private_pred'] = kmeans.labels_
```

```
#compare the results by scatterplot
sns.lmplot('Room.Board','Grad.Rate',data=df, hue='Private',
           palette='coolwarm',size=6,aspect=2,fit_reg=False)
sns.lmplot('Room.Board','Grad.Rate',data=df, hue='private_pred',
           palette='coolwarm',size=6,aspect=2,fit_reg=False)
```
